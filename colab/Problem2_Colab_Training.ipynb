````python
import subprocess
import sys
from pathlib import Path

# --- edit this to the folder that CONTAINS problem2/ ---
REPO = Path("/content/NLU_assignment_2")

DATA = Path("/content/TrainingNames.txt")
OUT = Path("/content/problem2_colab_out")

assert (REPO / "problem2" / "colab_train_all.py").is_file(), f"Expected {REPO}/problem2/ — fix REPO"
assert DATA.is_file(), f"Upload {DATA} first"

cmd = [
    sys.executable,
    "-m",
    "problem2.colab_train_all",
    "--data",
    str(DATA),
    "--out",
    str(OUT),
    "--epochs",
    "40",
    "--batch-size",
    "32",
    "--models",
    "rnn,blstm,attn",
]
print(" ".join(cmd))
subprocess.check_call(cmd, cwd=str(REPO))
print("Done. Download:", OUT)
````
