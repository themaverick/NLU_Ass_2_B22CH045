### Train RNN, Prefix BLSTM, and RNN+Attention
Edit **`REPO`** if your project lives elsewhere.


```python
import subprocess
import sys
from pathlib import Path

REPO = Path("/content/NLU_assignment_2")
DATA = Path("/content/TrainingNames.txt")
OUT = Path("/content/problem2_colab_out")

assert (REPO / "problem2" / "colab_train_all.py").is_file(), (
    f"Missing {REPO}/problem2/. Upload the repo and set REPO."
)
assert DATA.is_file(), f"Missing {DATA}. Upload TrainingNames.txt or run optional generation cell."

cmd = [
    sys.executable,
    "-m",
    "problem2.colab_train_all",
    "--data",
    str(DATA),
    "--out",
    str(OUT),
    "--epochs",
    "40",
    "--batch-size",
    "32",
    "--lr",
    "0.002",
    "--embed-dim",
    "64",
    "--hidden-dim",
    "128",
    "--models",
    "rnn,blstm,attn",
]
print(" ".join(cmd))
subprocess.check_call(cmd, cwd=str(REPO))
print("\nDone. Download folder:", OUT)
```
